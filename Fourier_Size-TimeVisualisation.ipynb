```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import time

def perform_fourier_transform(data_size):
    # Generating random data
    data = np.random.rand(data_size)

    # Performing Fourier transform
    start_time = time.time()
    transformed_data = np.fft.fft(data)
    frequencies = np.fft.fftfreq(len(data))
    end_time = time.time()
    result_time = end_time - start_time

    return data_size, result_time
```

```python
sizes_to_test = [1000, 10000, 100000, 1000000, 100000000]
results = []

for size in sizes_to_test:
    size_val, exec_time = perform_fourier_transform(size)
    results.append({'Size': size_val, 'ExecutionTime': exec_time})

# Creating a DataFrame from the results
df = pd.DataFrame(results)

# Plotting the data
plt.figure(figsize=(8, 6))
plt.plot(df['Size'], df['ExecutionTime'], marker='o')
plt.xscale('log')  # Log scale for better visualization
plt.yscale('log')
plt.title('Execution Time vs. Data Size')
plt.xlabel('Data Size')
plt.ylabel('Execution Time (seconds)')
plt.grid(True)
plt.show()
```
